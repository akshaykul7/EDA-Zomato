```python
df= df.drop(['url','address','phone','dish_liked','reviews_list','menu_item'],axis = 1)
```

```python
df.rename(columns={'listed_in(type)':'listed_type','listed_in(city)':'city'},inplace=True)
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['location','rest_type','cuisines','approx_cost(for two people)'],axis=0)
```

```python
def rate_clean(value):
    if (value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).replace('/5','')
        return float(value)        
```

```python
df['rate']=df['rate'].apply(rate_clean)
```

```python
df.select_dtypes(include = 'object')
```

```python
df['approx_cost(for two people)']= df['approx_cost(for two people)'].str.replace(',','').astype('int')
```

```python
rest_types_others= df['rest_type'].value_counts()<1000
rest_types_others
```

```python
rest_type_less_than_1000 = df['rest_type'].value_counts()[rest_types_others]
rest_type_less_than_1000
```

```python
def club_rest_type(value):
    if (value in rest_type_less_than_1000):
        return 'other'
    else:
        return value
```
